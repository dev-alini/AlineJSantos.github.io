```python
import pandas as pd
```

```python
import matplotlib.pyplot as plt
```

1. Ler o arquivo CSV

```python
url = ("https://raw.githubusercontent.com/profivan-ai/cdb-Python/refs/heads/main/arquivos/vendas_loja.csv")
df = pd.read_csv(url)
```

```python
df.head()
```

**2. Tratar os valores nulos**

```python
df.isnull()
```

```python
df.isnull().sum()
```

```python
df.fillna(0, inplace=True)
```

```python
df.isnull().sum()
```

**3. Gerar três gráficos usando somente matplotlib**

**Gráfico de Barras mostrando o total de vendas por categoria.**

```python
vendas_categoria = df.groupby('Categoria')["Vendas"].sum()
plt.figure(figsize=(8,5))
plt.bar(vendas_categoria.index, vendas_categoria.values, color='royalblue')
plt.title('Total de Vendas por Categoria')
plt.xlabel('Categoria')
plt.ylabel('Total de Vendas (R$)')
plt.show()
```

**Gráfico de Pizza mostrando a proporção de vendas por região**

```python
vendas_regiao = df.groupby("Regiao")["Vendas"].sum()
plt.figure(figsize=(7,7))
plt.pie(vendas_regiao, labels=vendas_regiao.index, autopct="%1.1f%%,")
plt.title("Distribuição de Vendas por Região")
plt.show()
```

**Gráfico de Linhas mostrando a evolução das vendas ao longo dos meses**

```python
vendas_mes = df.groupby("Mes")["Vendas"].sum()
plt.figure(figsize=(9,5))
plt.plot(vendas_mes.index, vendas_mes.values, marker=0, color='orange')
plt.title("Evolução das Vendas ao Longo dos Meses")
plt.xlabel("Mês")
plt.ylabel("Total de Vendas (R$)")
plt.show()
```

**1. Ler o arquivo CSV**

```python
url2 = "https://raw.githubusercontent.com/profivan-ai/cdb-Python/refs/heads/main/arquivos/pacientes.csv"
df2 = pd.read_csv(url2)
```

```python
df2.head()
```

**2. Tratar os valores nulos de forma adequada**

```python
df2.isnull().sum()
```

```python
df2.fillna(0, inplace=True)
```

```python
df2.isnull().sum()
```

3. Gerar os seguintes gráficos usando apenas matplotlib:
***

**Barras: média de glicose por faixa etária**

```python
df2.columns = df2.columns.str.strip()
```

```python
print(df2.head())
print(df2.columns.tolist())
```

**Gerar os seguintes gráficos usando apenas matplotlib:**

**Barras: média de glicose por faixa etária**

**Criar Faixa Etária**

```python

bins = [0, 18, 30, 45, 60, 120]
labels = ['0-17', '18-29', '30-44', '45-59', '60+']


df2["FaixaEtaria"] = pd.cut(df2["Idade"], bins=bins, labels=labels, right=False)

print(df2[["Idade", "FaixaEtaria"]].head())
```

```python
media_glicose = df2.groupby("FaixaEtaria")["Glicose"].mean()

plt.figure(figsize=(8,5))
plt.bar(media_glicose.index.astype(str), media_glicose.values, color="skyblue", edgecolor="black")
plt.title("Média de Glicose por Faixa Etária")
plt.xlabel("Faixa Etária")
plt.ylabel("Média de Glicose")
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.show()
```

**Pizza: proporção de pacientes por gênero**

```python
genero_contagem = df2["Genero"].value_counts()

plt.figure(figsize=(6,6))
plt.pie(genero_contagem, labels=genero_contagem.index, autopct='%1.1f%%', startangle=90, colors=["lightcoral", "lightskyblue", "lightgreen"])
plt.title("Distribuição de Pacientes por Gênero")
plt.show()
```

**Linhas: evolução média da pressão arterial por mês**

```python

ordem_meses = ["Jan", "Fev", "Mar", "Abr", "Mai", "Jun", "Jul", "Ago", "Set", "Out", "Nov", "Dez"]


df2["Mes"] = pd.Categorical(df2["Mes"], categories=ordem_meses, ordered=True)


media_pressao_mes = df2.groupby("Mes")["Pressao"].mean()


plt.figure(figsize=(8,5))
plt.plot(media_pressao_mes.index, media_pressao_mes.values, marker='o', linestyle='-', color='orange')
plt.title("Evolução Média da Pressão Arterial por Mês")
plt.xlabel("Mês")
plt.ylabel("Pressão Média")
plt.grid(True, linestyle='--', alpha=0.6)
plt.show()
```
